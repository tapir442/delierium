```python
from delierium.Infinitesimals import prolongationODE, overdeterminedSystemODE, Janet_Basis_from_ODE
from delierium.MatrixOrder import Context
from IPython.display import Math
from delierium.helpers import latexer
from delierium.JanetBasis import FindIntegrableConditions, _Differential_Polynomial
```

```python
x, y= var("x y")
xi  = function("xi")(x,y)
eta = function("eta")(x,y)
ctx = Context([xi, eta], [x,y])
xi_x = diff(xi, x)
xi_y = diff(xi, y)
xi_xx = diff(xi, x,x)
xi_yy = diff(xi, y,y)
xi_xy = diff(xi, x,y)
eta_x = diff(eta, x)
eta_y = diff(eta, y)
eta_xx = diff(eta, x,x)
eta_yy = diff(eta, y,y)
eta_xy = diff(eta, x,y)

A = function("A")(x,y)
B = function("B")(x,y)
C = function("C")(x,y)
D = function("D")(x,y)
A_x = diff(A, x)
A_y = diff(A, y)
B_x = diff(B, x)
B_y = diff(B, y)
C_x = diff(C, x)
C_y = diff(C, y)
D_x = diff(D, x)
D_y = diff(D, y)
```

```python
e1 = xi_yy - 2*A*eta_y - B*xi_y + A*xi_x - A_y*eta-A_x*xi
k1 = _Differential_Polynomial(e1, ctx)
display(Math(latexer(k1.expression())))
```

```python
e2 = eta_xx - D*eta_y +C*eta_x +2*D*xi_x +D_y*eta+D_x*xi
```

```python
e3 = eta_xy - xi_xx/2 +B*eta_x+3*D*xi_y/2+C*xi_x/2 + C_y*eta/2 + C_x*xi/2
```

```python
e4 = eta_yy - 2*xi_xy +B*eta_y + 3*A*eta_x + 2*C*xi_y + B_y*eta +B_x*xi
```

```python
kikuruz=FindIntegrableConditions([_Differential_Polynomial(_, ctx) for _ in [e1,e2,e3,e4]], ctx)
```
