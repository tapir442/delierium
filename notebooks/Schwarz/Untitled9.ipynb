```python
def expr_eq(e1, e2, vars):
    import random
    l = len(vars)
    i = 0
    rlist= []
    while i != l:
        r = random.randint(-1_000_000_000, 1_000_000_000)
        if r:
            rlist.append(r)
            i += 1
    r = dict(zip(vars, rlist))
    ev1 = e1.subs(r)
    ev2 = e2.subs(r)
    return bool(ev1 == ev2)
```

```python
def is_zero(e, vars):
    import random
    l = len(vars)
    i = 0
    rlist= []
    while i != l:
        r = random.randint(-1_000_000_000, 1_000_000_000)
        if r:
            rlist.append(r)
            i += 1
    ev = e.subs(dict(zip(vars, rlist)))
    return bool(ev == 0)
```
